```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get('https://www.cnbc.com/world/?region=world')
page

soup = BeautifulSoup(page.content)
soup

headline = []
for i in soup.find_all('div',class_="RiverHeadline-headline RiverHeadline-hasThumbnail"):
    headline.append(i.text)
headline

news = []
for i in soup.find_all('span',class_="QuickLinks-quickLink"):
    news.append(i.text)
news


time = []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
time


df2 = pd.DataFrame({'Headline': headline, })
print(df2)

df3 = pd.DataFrame({'NewsLink': news  })
print(df3)

df4 = pd.DataFrame({'Time': time  })
print(df4)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for Men's ODI rankings
url_men = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"

# Send a GET request to the URL
response = requests.get(url_men)

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, 'html.parser')

# Scrape the data for the top 10 ODI teams
teams = []
matches = []
points = []
ratings = []

table = soup.find('table', {'class': 'table'})
rows = table.find_all('tr')[1:11]

for row in rows:
    cols = row.find_all('td')
    teams.append(cols[1].text.strip())
    matches.append(cols[2].text.strip())
    points.append(cols[3].text.strip())
    ratings.append(cols[4].text.strip())

# Create a DataFrame for Men's ODI teams
data_men = {
    "Team": teams,
    "Matches": matches,
    "Points": points,
    "Rating": ratings
}

df_men = pd.DataFrame(data_men)

# Display the data frame
print("Top 10 Men's ODI Teams:")
print(df_men)
```
